```python
import numpy as np
import statistics as st
import random
import matplotlib.pyplot as plt
```

```python
X = np.random.random((200, 1))
X[:10]
```

```python
print(f"Mean of X: {np.mean(X)}\nStandard Deviation of X: {np.std(X)}")
```

```python
# Normalization

def Normalize(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))

XN = Normalize(X)
XN[:10]
```

```python
print(f"Mean of XN: {np.mean(XN)}\nStandard Deviation of XN: {np.std(XN)}")
```

```python
# Standardization

def Standardize(data):
    return (data - np.mean(data)) / np.std(data)

XS = Standardize(X)
XS[:10]
```

```python
# Plotting (Data - Normalized)

fig, ax = plt.subplots(1, 3, figsize=(20, 5))

ax[0].plot(
    X,
    "^",
    color = "blue",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Randomly Generated Data",
);
ax[0].set_title("Randomly Generated Data", fontsize = 12);
ax[0].legend(loc = "upper right", fontsize = 8);

ax[1].plot(
    XN,
    "v",
    color = "red",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Normalized Data"
);
ax[1].set_title("Normalized Randomly Generated Data", fontsize = 12);
ax[1].legend(loc = "upper right", fontsize = 8);

ax[2].plot(
    X,
    "^",
    color = "blue",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Randomly Generated Data"
);
ax[2].plot(
    XN,
    "v",
    color = "red",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Standardized Data"
);

ax[2].set_title("Normalized & Randomly Generated Data Together", fontsize = 12);
ax[2].legend(loc = "upper right", fontsize = 8);

X_var = [X, XN]
for i in range(len(X_var)):
    ax[i].plot(
        np.std(X_var[i]),
        "o",
        alpha = 0.8,
        color = "red",
        markersize = 7,
        markeredgecolor = "black",
);
```

```python
# Plotting (Data - Standardized)

fig, ax = plt.subplots(1, 3, figsize=(20, 5))

ax[0].plot(
    X,
    "^",
    color = "blue",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Randomly Generated Data"
);
ax[0].set_title("Randomly Generated Data", fontsize = 12);
ax[0].legend(loc = "upper right", fontsize = 8);

ax[1].plot(
    XS,
    "v",
    color = "green",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Standardized Data"
);
ax[1].set_title("Standardized Data", fontsize = 12);
ax[1].legend(loc = "upper right", fontsize = 8);

ax[2].plot(
    X,
    "^",
    color = "blue",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Randomly Generated Data"
);
ax[2].plot(
    XS,
    "v",
    color = "green",
    alpha = 0.5,
    markersize = 5,
    markeredgecolor = "black",
    label = "Standardized Data"
);

ax[2].set_title("Standardized & Randomly Generated Data Together", fontsize = 12);
ax[2].legend(loc = "upper right", fontsize = 8);

X_var = [X, XS]
for i in range(len(X_var)):
    ax[i].plot(
        np.std(X_var[i]),
        "o",
        alpha = 0.8,
        color = "red",
        markersize = 7,
        markeredgecolor = "black",
);
```

### Scatter Matrix

```python
import pandas as pd

df_random = pd.DataFrame(np.random.randn(1000, 4), columns=['A', 'B', 'C', 'D'])
df_random.head()
```

```python
pd.plotting.scatter_matrix(
    frame = df_random,
    alpha = 0.5,
    figsize = (10, 10),
    grid = True,
    marker = "o",
    range_padding = 0.1,
);
```

### Correlations

```python
corr_random = df_random.corr(
    method = "spearman",
    min_periods = 1
)
corr_random.style.background_gradient(cmap = 'coolwarm')
```
